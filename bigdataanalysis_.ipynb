## 연관 분석  
장바구니 구매 물품 데이터셋으로 물품들 간의 연관관계를 파악하고 성능을 평가해본다.

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/retail_dataset.csv', sep=',')
```

```python
df
```

```python
df.info()
```

```python
# 장바구니 목록 값을 수치로 표현하기 위하여 각 항목당 매칭될 경우 1로, 아니면 0으로 표시한다.
items = set()
for col in df:
  items.update((df[col].unique()))

itemset = set(items)
encoding = []
for index, row in df.iterrows():
  rowset = set(row)
  labels = {}
  dismatching = list(itemset - rowset)
  matching = list(itemset.intersection(rowset))
  for item in dismatching:
    labels[item] = 0
  for item in matching:
    labels[item] = 1
  encoding.append(labels)

encoding[0]
result = pd.DataFrame(encoding)
result
```

```python
# 첫째 Nan 열항목 삭제 처리하기
result = result.drop(result.columns[0], axis=1)
result
```

```python
# mixtend의 apiriori 사용
freq_items= apriori(result, min_support=0.2, use_colnames=True)
```

```python
# 연관 규칙 도출(신뢰도 임계치 0.6 기반)
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules.head()
```

# 통계적 가설 검정 기법 실습

```python
from logging import critical
from scipy.stats import norm
#표본의 평균 검정
data = np.array([23, 25, 28, 30, 26, 27, 29, 32, 31, 28])

#모집단의 평균과 분산(표준편차)
population_mean =  28 # 모집단의 평균
population_variance = 4 # 모집단의 분산(표준편차 제곱)

#유의수준 결정
alpha = 0.05

# 단측 또는 양측 검정 선택(예:양측 검정)
alternative = "two.sided"

# 표준평균 계산
sample_mean = np.mean(data)

# 표본 크기
n = len(data)

# 검정통계량 계산 (Z-검정)
test_staticstic = (sample_mean - population_mean)/(np.sqrt(population_variance)/np.sqrt(n))

# 임계값 계산(z-검정, 양측 검정)
critical_value = norm.ppf(1 - alpha / 2)

# p-value 계산
p_value = 2 * (1-norm.cdf(abs(test_staticstic)))

print("표본평균:", sample_mean)
print("검정통계량: ", test_staticstic)
print("임계값: ", critical_value)
print("p-값:", p_value)

if abs(test_staticstic) > critical_value:
  print("귀무가설 기각")
else:
  print("귀무가설 채택")
```

```python
# cut-off 정의
def cut_off(y, threshold):
  Y = y.copy()
  Y[Y>threshold] = 1
  Y[Y<=threshold] = 0
  return(Y.astype(int))
```
